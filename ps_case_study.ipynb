```python
import numpy as np
import matplotlib.pyplot as plt

engine_cc = np.array([125,350,110,110,125,145,145])
mileage = np.array([55,30,60,65,50,45,65])
n = np.size(engine_cc)
x_bar = np.mean(engine_cc)
y_bar = np.mean(mileage )
cov = sum((engine_cc - x_bar) * (mileage - y_bar)) / n
v1 = np.var(engine_cc)
v2 = np.var(mileage )
Correlation = cov / np.sqrt(v1 * v2)
print("Correlation:", round(Correlation, 3))

```

```python
import numpy as np
import matplotlib.pyplot as plt

engine_cc = np.array([125,350,110,110,125,145,145])
mileage = np.array([55,30,60,65,50,45,65])

coefficients = np.polyfit(engine_cc, mileage, 2)
a, b, c = coefficients
a = np.round(a,6)
b = np.round(b,6)
c = np.round(c,6)
print(" Quadratic Equation: {}x^2 + {}x + {}".format(a, b, c))

engine_cc_range = np.linspace(min(engine_cc), max(engine_cc), 100)

mileage_curve = a * engine_cc_range**2 + b * engine_cc_range + c

plt.scatter(engine_cc, mileage, label="Data", color="blue")
plt.plot(engine_cc_range, mileage_curve, label="Fitted Curve", color="red")
plt.xlabel("Engine CC")
plt.ylabel("Mileage")
plt.legend()
plt.title("Quadratic Curve")
plt.grid(True)
plt.show()
```
